```python
failas_2021 = 'ei_2021_12_31.csv'
df_2021 = pd.read_csv(failas_2021)

failas_2016 = 'ei_2016_12_31.json'
df_2016 = pd.read_json(failas_2016)

df_2021['Year'] = [pd.to_datetime(date).year for date in df_2021['registravimoData']]
df_2016['Year'] = [pd.to_datetime(date).year for date in df_2016['registravimoData']]

df_sujungta = pd.concat([df_2021, df_2016])

sujungti_duomenys = df_sujungta.groupby(['Year', 'schema1']).size().unstack(fill_value=0)

sujungti_duomenys.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Ivykiu skaicius pagal metus')
plt.xlabel('Metai')
plt.ylabel('Ivykiu skaicius')
plt.legend(title='Ivykio rusis', bbox_to_anchor= (1, 1))

plt.show()
```

Matom, kad eismo ivykiu padaugejo 2021 metais. Daugiausia buvo nuvaziavimai nuo kelio, susidurimai judant ta pacia kryptimi. Panasus ivykiu rusiu kiekis abejais metais

```python
df_2021_zuvusieji = df_2021.groupby('Year')['zuvusiuSkaicius'].sum()
df_2016_zuvusieji = df_2016.groupby('Year')['zuvusiuSkaicius'].sum()

df_sujungti_zuvusieji = pd.DataFrame({
    '2021':df_2021_zuvusieji,
    '2016':df_2016_zuvusieji
})

plt.figure(figsize=(10,6))
plt.pie(df_sujungti_zuvusieji.sum(), labels=df_sujungti_zuvusieji.columns, autopct='%1.1f%%', startangle=90, colors=['blue','red'])
plt.title('Kuriais metais buvo daugiau zuvusiu')

plt.show()
```

Matom, kad zuvusiu kiekis yra didesnis 2016 metais, nors paciu eismo ivykiu buvo maziau

```python
df_2021['suzeistuSkaicius'] = pd.to_numeric(df_2021['suzeistuSkaicius'])
df_2016['suzeistuSkaicius'] = pd.to_numeric(df_2016['suzeistuSkaicius'])

suzeisti_zmones_2021 = df_2021['suzeistuSkaicius'].sum()
suzeisti_zmones_2016 = df_2016['suzeistuSkaicius'].sum()

plt.bar(['2021', '2016'], [suzeisti_zmones_2021, suzeisti_zmones_2016])
plt.title('Bendras suzeistu zmoniu skaicius skirtingais metais')
plt.xlabel('Metai')
plt.ylabel('Suzeisti zmones')

plt.show()
```

Ta pati matom ir su suzeistais zmonemis. Nors ir ivykiu buvo maziau 2016, taciau ir mirciu ir suzeidimu daugiau

```python
df_2021['suzeistuSkaicius'] = pd.to_numeric(df_2021['suzeistuSkaicius'])
df_2021['suzeistaVaiku'] = pd.to_numeric(df_2021['suzeistaVaiku'])

suzeisti_zmones = df_2021['suzeistuSkaicius'].sum()
suzeisti_vaikai = df_2021['suzeistaVaiku'].sum()

suzeistu_vaiku_procentas = (suzeisti_vaikai/suzeisti_zmones)*100
suzeistu_ne_vaiku_procentas = 100 - suzeistu_vaiku_procentas

plt.figure(figsize=(5,5))
plt.pie([suzeistu_vaiku_procentas, suzeistu_ne_vaiku_procentas], labels=['Suzeisti vaikai', 'Suzeisti, bet ne vaikai'], autopct='%1.1f%%')
plt.title('Suzeistu vaiku procentas 2021 metais')

plt.show()
```

```python
df_2016['suzeistuSkaicius'] = pd.to_numeric(df_2016['suzeistuSkaicius'])
df_2016['suzeistaVaiku'] = pd.to_numeric(df_2016['suzeistaVaiku'])

suzeisti_zmones = df_2016['suzeistuSkaicius'].sum()
suzeisti_vaikai = df_2016['suzeistaVaiku'].sum()

suzeistu_vaiku_procentas = (suzeisti_vaikai/suzeisti_zmones)*100
suzeistu_ne_vaiku_procentas = 100 - suzeistu_vaiku_procentas

plt.figure(figsize=(5,5))
plt.pie([suzeistu_vaiku_procentas, suzeistu_ne_vaiku_procentas], labels=['Suzeisti vaikai', 'Suzeisti, bet ne vaikai'], autopct='%1.1f%%')
plt.title('Suzeistu vaiku procentas 2016 metais')

plt.show()
```

Matom, kad nors ir suzeidimu buvo maziau 2021 metais, taciau is ju didesne dali sudaro vaikai negu 2016 metais

```python
df_2021['suzeistaVaiku'] = pd.to_numeric(df_2021['suzeistaVaiku'])
df_2016['suzeistaVaiku'] = pd.to_numeric(df_2016['suzeistaVaiku'])

suzeisti_vaikai_2021 = df_2021['suzeistaVaiku'].sum()
suzeisti_vaikai_2016 = df_2016['suzeistaVaiku'].sum()

plt.bar(['2021','2016'], [suzeisti_vaikai_2021, suzeisti_vaikai_2016], color=['blue','red'])
plt.title('Kiek buvo suzeistu vaiku')
plt.xlabel('Metai')
plt.ylabel('Suzeisti vaikai')

plt.show()
```
